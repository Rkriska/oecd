```python
# ===============================================
# OECD-NEA-1 – Full Dummy Pipeline
# ===============================================

import os
import sys
import importlib
import pandas as pd
from concurrent.futures import ThreadPoolExecutor

# -------------------------------
# 1. Tambahkan debug/ ke path
# -------------------------------
debug_path = os.path.join(os.getcwd(), "debug")
if debug_path not in sys.path:
    sys.path.append(debug_path)

# -------------------------------
# 2. Import modules
# -------------------------------
import extract_excel
import pipeline
import few_shot_builder
import importlib

importlib.reload(extract_excel)
importlib.reload(pipeline)
importlib.reload(few_shot_builder)

# -------------------------------
# 3. Load semua Excel dari inputs/
# -------------------------------
input_folder = "inputs"
dfs = extract_excel.extract_all_excels(input_folder)

if not dfs:
    raise ValueError("Tidak ada Excel yang berhasil dibaca dari inputs/!")

for fname, df in dfs.items():
    print(f"{fname}: {df.shape[0]} rows x {df.shape[1]} cols")

# -------------------------------
# 4. Ambil sample row pertama → test single row
# -------------------------------
first_df = None
first_file = None
for fname, df in dfs.items():
    if len(df) > 0:
        first_df = df
        first_file = fname
        break

sample_text = extract_excel.format_df_to_llm_text(first_df, row_index=0)[0]
print(f"\n=== Sample Input Row from {first_file} ===")
print(sample_text[:500], "\n")

# -------------------------------
# 5. Test pipeline single row (dummy)
# -------------------------------
result = pipeline.process_single_risk(sample_text)
print("=== Pipeline Single Row Output (Dummy) ===")
for k, v in result.items():
    print(f"{k}: {v}")

# -------------------------------
# 6. Generate few-shot JSON contoh
# -------------------------------
json_risk_id = few_shot_builder.get_few_shots_for_column("Risk ID")
print("\nFew-Shot Example (Risk ID):")
print(json_risk_id)

# -------------------------------
# 7. Batch processing semua row (dummy)
# -------------------------------
all_results = []

for fname, df in dfs.items():
    for idx in range(len(df)):
        row_text = extract_excel.format_df_to_llm_text(df, row_index=idx)[0]
        try:
            res = pipeline.process_single_risk(row_text)
        except:
            res = {col: "DUMMY" for col in [
                "Risk ID","Risk Description","Project Stage","Project Category",
                "Risk Owner","Mitigating Action","Likelihood (1-10)","Impact (1-10)","Risk Priority"
            ]}
        res['Source File'] = fname
        res['Row Index'] = idx
        all_results.append(res)

batch_file = "pipeline_results_dummy.xlsx"
pd.DataFrame(all_results).to_excel(batch_file, index=False)
print(f"\nBatch processing done → '{batch_file}'")

# -------------------------------
# 8. Generate final outputs → Excel + JSON
# -------------------------------
df_final = pd.read_excel(batch_file)

# Postprocessing dummy
for col in ["Likelihood (1-10)", "Impact (1-10)"]:
    df_final[col] = pd.to_numeric(df_final[col], errors='coerce')
    df_final[col] = df_final[col].fillna(5).clip(1,10)

df_final["Risk Priority"] = df_final["Risk Priority"].apply(lambda x: x if str(x) in ["Low","Med","High"] else "Med")

# Save Excel final
excel_output = "final_outputs.xlsx"
df_final.to_excel(excel_output, index=False)
print(f"Final Excel saved → '{excel_output}'")

# Save JSON final
json_output = "final_outputs.json"
df_final.to_json(json_output, orient="records", indent=2)
print(f"Final JSON saved → '{json_output}'")
```

```python
import tiktoken
import extract_excel
import os

# -------------------------------
# Pilih model LLM untuk token count
# -------------------------------
model_name = "gpt-4"  # gpt-4 atau deepseek-chat
enc = tiktoken.encoding_for_model(model_name)

# -------------------------------
# Load semua Excel
# -------------------------------
input_folder = "inputs"
dfs = extract_excel.extract_all_excels(input_folder)

# -------------------------------
# Ambil sample beberapa row untuk estimasi
# -------------------------------
sample_texts = []
for fname, df in dfs.items():
    for idx in range(min(len(df), 5)):  # ambil max 5 row per file untuk estimasi
        row_text = extract_excel.format_df_to_llm_text(df, row_index=idx)[0]
        sample_texts.append(row_text)

# -------------------------------
# Hitung token
# -------------------------------
def count_tokens(text, enc):
    return len(enc.encode(text))

token_estimates = [count_tokens(t, enc) for t in sample_texts]

for i, t_count in enumerate(token_estimates):
    print(f"Row {i+1}: {t_count} tokens")

print(f"\nAverage tokens per row: {sum(token_estimates)/len(token_estimates):.1f}")
```

```python
# ===============================================
# OECD-NEA-1 – Single Row Test Hugging Face
# ===============================================

from transformers import pipeline
import extract_excel
import importlib

# Reload module
importlib.reload(extract_excel)

# -------------------------------
# Load Excel → ambil sample row pertama
# -------------------------------
input_folder = "inputs"
dfs = extract_excel.extract_all_excels(input_folder)

first_df = None
first_file = None
for fname, df in dfs.items():
    if len(df) > 0:
        first_df = df
        first_file = fname
        break

sample_text = extract_excel.format_df_to_llm_text(first_df, row_index=0)[0]
print(f"\nSample Row Text from {first_file} (preview 500 chars):")
print(sample_text[:500], "...")

# -------------------------------
# Setup Hugging Face pipeline
# -------------------------------
# Bisa pilih model HF gratis / open-source
model_name = "tiiuae/falcon-7b-instruct"  # contoh gratis
hf_pipeline = pipeline("text2text-generation", model=model_name, device=-1)  # device=-1 = CPU

# -------------------------------
# Run single row
# -------------------------------
prompt = f"""
You are a risk management expert. Extract 9 columns from this risk register row: 
Risk ID, Risk Description, Project Stage, Project Category, Risk Owner, 
Mitigating Action, Likelihood (1-10), Impact (1-10), Risk Priority.

Input Row:
{sample_text}
"""

hf_output = hf_pipeline(prompt, max_new_tokens=512, do_sample=False)
print("\n=== Hugging Face Output ===")
print(hf_output[0]['generated_text'])
```
